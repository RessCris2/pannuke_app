```python
from pycocotools.coco import COCO
from matplotlib import pyplot as plt
```

```python
def desc_datasets(ann_file):
    coco = COCO(ann_file)
    anns = coco.anns
    
    # 每张图片有多少细胞核？
    img_cnt = {}
    img_area = {}
    # 每个类别有多少细胞核？
    cate_cnt = {}
    for idx in anns:
        ann = anns[idx]
        if ann['category_id'] in cate_cnt:
            cate_cnt[ann['category_id']] += 1
        else:
            cate_cnt[ann['category_id']]=1
            
        if ann['image_id'] in img_cnt:
            img_cnt[ann['image_id']] += 1
            img_area[ann['image_id']] += ann['area']
        else:
            img_cnt[ann['image_id']]=1
            img_area[ann['image_id']] = ann['area']
        
    
    avg_area = {}
    for image in img_cnt:
        avg_area[image] = img_area[image]/img_cnt[image]
    # 一共多少图片
    imgs_len = len(coco.imgs)
    # 一共多少细胞核
    nuclei_len = len(coco.anns)
    # 细胞核的平均大小？
    mean_area = sum(img_area.values())/sum(img_cnt.values())
    print(f"""数据集的情况如下：
            共有图片{imgs_len}张，
            细胞核{nuclei_len}个, 
            细胞核平均大小为{mean_area}
             """)
    return img_cnt, cate_cnt, avg_area, imgs_len, nuclei_len,  mean_area
```

```python
ann_file = "/root/autodl-tmp/pannuke_app/datasets/processed/PanNuke/test/test_annotations.json"
```

```python
img_cnt, cate_cnt, avg_area, imgs_len, nuclei_len,  mean_area = desc_datasets(ann_file)
```
